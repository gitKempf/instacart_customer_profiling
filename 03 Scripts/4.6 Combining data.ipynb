# 4.6 Merging and exporting data.

```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# Reletive path
path = '../'

# 3. Loading dataframe orders_products_combined.pkl
orders_products_combined = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_combined.pkl'))
```

```python
# 4. Check the shape of the imported dataframe (it should be the same as the one you exportedâ€”always check!).
orders_products_combined.shape
```

```python
orders_products_combined.head()
```

```python
# Removing old _merge column to generate new one
orders_products_combined.drop('_merge', axis=1, inplace=True)
```

```python
# Loading dataframe products_cleaned.csv
df_products = pd.read_csv(
    os.path.join(path, '02 Data', 'Prepared Data', 'products_cleaned.csv')
)
```

```python
df_products.head()
```

```python
# Mergeing the order_products__prior dataframe with the df_ords dataframe
orders_products_merged = orders_products_combined.merge(df_products, on = 'product_id', how = 'outer', indicator = True)
```

```python
orders_products_merged.head()
```

```python
# 6. Confirm the results of the merge using the merge flag.
orders_products_merged['_merge'].value_counts()
```

```python
# 7. Export the newly created dataframe as orders_products_merged in a suitable format (taking into consideration the size).
orders_products_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
