```python
# Aggrageting a mean order number grouped by a department
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Alternative way of Aggrageting a mean order number grouped by a department 
df.groupby('department_id').order_number.mean()
```

```python
# Producing not only the mean but also the min and max of the “order_number” column for df dataframe
df_stat = df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
df_stat
```

```python
# Producing the mean, the min and the max of the “order_number” column for ords_prods_merge dataframe
ords_prods_merge_stat = ords_prods_merge.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
ords_prods_merge_stat
```

```python
compared_stats = ords_prods_merge_stat.merge(df_stat, on = 'department_id', how = 'left')
```

```python
# Comparing the mean values in the subset an in the full dataframe
compared_stats['diff'] = (1 - compared_stats['order_number_x']['mean'] / compared_stats['order_number_y']['mean']) * 100
compared_stats['diff']
```

## 4. Loyalty Flag

```python
# Creating a new column containing the maximum frequency of the “order_number” column.
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)

ords_prods_merge.head(10)
```

```python
# Using the loc() function, which will create a second column containing a flag designating whether a customer is “loyal” or not.
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(10)
```

## 5. Basic statistics of the product prices for each loyalty category

```python
# Checking statistics of prices by customer loyalty
ords_prods_merge_product_stat_by_loyalty = \
    ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})

ords_prods_merge_product_stat_by_loyalty.head()
```

Observation: Surprisingly, the average product prices paid by loyal customers are slightly lower than those paid by regular customers and new customers.

## 6. Customers spending labeling

```python
# Creating a new column containing the mean price paid by customer.
ords_prods_merge['user_mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)

ords_prods_merge.head(10)
```

```python
# Using the loc() function, which will create a second column, 
# containing a flag designating whether a customer is “High spender” or not.
ords_prods_merge.loc[ords_prods_merge['user_mean_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['user_mean_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.head()
```

```python
# Check the number of users in spending groups
ords_prods_merge_grouped_by_users = ords_prods_merge.groupby(['user_id'])['spending_flag'].first()
ords_prods_merge_grouped_by_users.value_counts(dropna = False)
```

## 7. Orders frequency

```python
# creating “order_frequency_flag” column using transform()
ords_prods_merge['median_prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Using the loc() function, which will create a second column, 
# containing a flag designating whether a customer is “Frequent customer” or not.
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'

ords_prods_merge.loc[(ords_prods_merge['median_prior_orders'] > 10) \
                              & (ords_prods_merge['median_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'

ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'

ords_prods_merge.head()
```
