### 1. Product ranges

```python
# Reducing the leangth
df = ords_prods_merged[:1000000]
```

```python
# Defining function to lable the products with product ranges.
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Allpying labling function to products in dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Checking ranges
df['price_range'].value_counts()
```

```python
df['prices'].max()
```

### 2. Product ranges with loc()

```python
# Labling the subset of products with product ranges
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Labling the all of products with product ranges
ords_prods_merged.loc[ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
ords_prods_merged.loc[(ords_prods_merged['prices'] <= 15) & (ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
ords_prods_merged.loc[ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

```python
# 2. Fixing “Busiest day” to become “Busiest days” (plural). 
# This label should correspond with the two busiest days of the week as opposed to the single busiest day. 
# Adding two slowest days to logic
# Labling busy days

result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value in [0,1]:
    result.append("Busiest day")
  elif value in [4,4]:
    result.append("Least busy")
  else:
    result.append("Regularly busy")

# 3. Creating column with busiest_days. Checking the values of this new column for accuracy.
# Removing old _merge column to generate new one
ords_prods_merged.drop('busiest_day', axis=1, inplace=True)
ords_prods_merged["busiest_days"] = result
ords_prods_merged["busiest_days"].value_counts(dropna = False)
```

### Question 4. Busiest hours

```python
# When too many users make Instacart orders at the same time, the app freezes. 
# The senior technical officer at Instacart wants to identify the busiest hours of the day. 


busiest_hours = ords_prods_merged['order_hour_of_day'].value_counts()
busiest_hours
```

```python
# Converting series to list to eseare subset indexes
busiest_hours_index = list(busiest_hours.index)
```

```python
# Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” 
# Creating a new column containing these labels called “busiest_period_of_day.”
ords_prods_merged.loc[
    ords_prods_merged['order_hour_of_day'].isin(busiest_hours_index[:7]), 'busiest_period_of_day'
] = 'Most orders'

ords_prods_merged.loc[
    ords_prods_merged['order_hour_of_day'].isin(busiest_hours_index[7:15]), 'busiest_period_of_day'
] = 'Average orders' 

ords_prods_merged.loc[
    ords_prods_merged['order_hour_of_day'].isin(busiest_hours_index[15:]), 'busiest_period_of_day'
] = 'Fewest orders'

ords_prods_merged.loc[
    ~ords_prods_merged['order_hour_of_day'].isin(busiest_hours_index), 'busiest_period_of_day'
] = 'Not enough data'

# 5. Printing the frequency for this new column.
ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
